### CART 和决策树的超参数

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets
X, y = datasets.make_moons(noise = 0.25, random_state=666)
```

```python
plt.scatter(X[y==0, 0], X[y==0,1])
plt.scatter(X[y==1, 0], X[y==1,1])
plt.show
```

```python
def plot_decision_boundary(model, axis):
  x0, x1 = np.meshgrid(
    np.linspace(axis[0], axis[1], int((axis[1] - axis[0]) * 100)),
    np.linspace(axis[2], axis[3], int((axis[3] - axis[2]) * 100))
  )
  X_new = np.c_[x0.ravel(), x1.ravel()]
  
  y_predict = model.predict(X_new)
  zz = y_predict.reshape(x0.shape)
  
  from matplotlib.colors import ListedColormap
  custom_cmap = ListedColormap(['#EF9A9A', '#FFF59D', '#90CAF9'])
  
  plt.contourf(x0, x1, zz, cmap=custom_cmap)

from sklearn.tree import DecisionTreeClassifier 
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X, y)
```

```python
dt_classifier.get_params()
```

```python
plot_decision_boundary(dt_classifier, axis=[-1.5, 2.5, -1.0,1.5])

plt.scatter(X[y==0, 0], X[y==0,1])
plt.scatter(X[y==1, 0], X[y==1,1])
# plt.scatter(X[y==2, 0], X[y==2,1])
plt.show()
```

```python
dt_classifier = DecisionTreeClassifier(max_depth=2)
dt_classifier.fit(X, y)
plot_decision_boundary(dt_classifier, axis=[-1.5, 2.5, -1.0,1.5])

plt.scatter(X[y==0, 0], X[y==0,1])
plt.scatter(X[y==1, 0], X[y==1,1])
# plt.scatter(X[y==2, 0], X[y==2,1])
plt.show()
```

```python
dt_classifier = DecisionTreeClassifier(min_samples_split=10)
dt_classifier.fit(X, y)
plot_decision_boundary(dt_classifier, axis=[-1.5, 2.5, -1.0,1.5])

plt.scatter(X[y==0, 0], X[y==0,1])
plt.scatter(X[y==1, 0], X[y==1,1])
# plt.scatter(X[y==2, 0], X[y==2,1])
plt.show()
```

```python
dt_classifier = DecisionTreeClassifier(min_samples_split=100)
dt_classifier.fit(X, y)
plot_decision_boundary(dt_classifier, axis=[-1.5, 2.5, -1.0,1.5])

plt.scatter(X[y==0, 0], X[y==0,1])
plt.scatter(X[y==1, 0], X[y==1,1])
# plt.scatter(X[y==2, 0], X[y==2,1])
plt.show()
```

```python
dt_classifier = DecisionTreeClassifier(min_samples_leaf=2)
dt_classifier.fit(X, y)
plot_decision_boundary(dt_classifier, axis=[-1.5, 2.5, -1.0,1.5])

plt.scatter(X[y==0, 0], X[y==0,1])
plt.scatter(X[y==1, 0], X[y==1,1])
# plt.scatter(X[y==2, 0], X[y==2,1])
plt.show()
```

```python
dt_classifier = DecisionTreeClassifier(max_leaf_nodes=8)
dt_classifier.fit(X, y)
plot_decision_boundary(dt_classifier, axis=[-1.5, 2.5, -1.0,1.5])

plt.scatter(X[y==0, 0], X[y==0,1])
plt.scatter(X[y==1, 0], X[y==1,1])
# plt.scatter(X[y==2, 0], X[y==2,1])
plt.show()
```
