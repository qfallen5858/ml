```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10, 100)
```

```python
y = np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.plot(x, y )
plt.show()
```

```python
cosy = np.cos(x)
siny = y.copy()
plt.plot(x, siny)
plt.plot(x, cosy)
plt.axis([-1,11,-2,2])
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
plt.plot(x, siny, label = "sin(x)")
plt.plot(x, cosy, label = "cos(x)")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.title('welcome')
plt.show()
```
